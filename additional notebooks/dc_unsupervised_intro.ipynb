```python
import numpy as np
import random
import matplotlib.pyplot as plt

plt.style.use('ggplot')
```

```python
a = np.array([random.randint(0, 500) for i in range(10)])
a
```

```python
x_bar = np.mean(a)
x_bar
```

```python
a - x_bar
```

```python
plt.figure(figsize = (4,4))
plt.hist(a, color='deepskyblue')
plt.axvline(x_bar, color='orangered')
plt.title("Original data")
plt.show()
```

```python
round(np.mean(a - x_bar)) # close to 0.
```

```python
plt.figure(figsize = (4,4))
plt.hist(a - x_bar, color='deepskyblue')
plt.axvline(0, color='orangered')
plt.title("x - mean")
plt.show()
```

```python
s = np.std(a)
s
```

```python
np.std(a - x_bar)
```

```python
np.std((a - x_bar)/s)
```

```python
round(np.mean((a - x_bar)/s))
```

```python
plt.figure(figsize = (4,4))
plt.hist((a - x_bar)/s, color='deepskyblue')
plt.axvline(0, color='orangered')
plt.title("(x - mean) / std")
plt.show()
```

```python
a_2d = np.reshape(a, (-1, 1))
a_2d
```

```python
from sklearn.preprocessing import normalize

normalized_a = normalize(a_2d, axis=0)
normalized_a
```

```python
plt.figure(figsize = (4,4))
plt.hist(normalized_a, color='deepskyblue')
plt.axvline(np.mean(normalized_a), color='orangered')
plt.title("normalized x (range [0,1])")
plt.show()
```
